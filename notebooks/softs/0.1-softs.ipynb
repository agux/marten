```python
from marten.utils.softs import SOFTSPredictor
import pandas as pd
import numpy as np
```

```python
seq_len = 5
```

```python
# mock a dataframe with time series stock market data, and randomized NaN values (missing values). 
# the date column shall be the first column (not index)
# missing data shall be sporadic, not necesarrily entire row.


# Generate date range
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')

# Generate random stock data
np.random.seed(42)
data = np.random.randn(100, 4) * 20 + 100

# Introduce NaN values randomly
nan_indices = np.random.choice(data.size, size=20, replace=False)
data.ravel()[nan_indices] = np.nan

# Create DataFrame
columns = ['Date', 'Open', 'High', 'Low', 'Close']
df = pd.DataFrame(data, columns=columns[1:])
df.insert(0, 'Date', dates)

df.head()
```
