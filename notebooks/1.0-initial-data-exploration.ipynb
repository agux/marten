```python
# 基金分红排行
# 接口: fund_fh_rank_em
# 目标地址: http://fund.eastmoney.com/data/fundleijifenhong.html
# 描述: 天天基金网-基金数据-分红送配-基金分红排行
# 限量: 单次返回所有历史数据
fund_fh_rank_em_df = ak.fund_fh_rank_em()
fund_fh_rank_em_df = fund_fh_rank_em_df[fund_fh_rank_em_df['基金简称'].str.contains('ETF')]
fund_fh_rank_em_df['average_dividend_per_share'] = fund_fh_rank_em_df['累计分红'] / fund_fh_rank_em_df['累计次数']
fund_fh_rank_em_df.sort_values(by='average_dividend_per_share', ascending=False, inplace=True)
print(fund_fh_rank_em_df)
# 累计分红： 注意单位: 元/份
```

```python
# 开放式基金排行
# 接口: fund_open_fund_rank_em
# 目标地址: https://fund.eastmoney.com/data/fundranking.html
# 描述: 东方财富网-数据中心-开放式基金排行
# 限量: 单次返回当前时刻所有数据
# symbol="全部"; choice of {"全部", "股票型", "混合型", "债券型", "指数型", "QDII", "LOF", "FOF"}
fund_open_fund_rank_em_df = ak.fund_open_fund_rank_em(symbol="全部")
print(fund_open_fund_rank_em_df)
```

## Rank by Fund Discount Rate

```python
df = ak.fund_etf_spot_em()
df["最新份额"] = df["最新份额"] / 100000000.  # 转换单位：亿
df["成交额"] = round(df["成交额"] / 100000000., 2)  # 转换单位：亿
df["主力净流入-净额"] = df["主力净流入-净额"] / 10000.0  # 转换单位：万
# filter df by `最新份额` > 1. and `涨跌幅` > 0.
# df = df[(df['最新份额'] > 1.) & (df['涨跌幅'] > 0.)]
df = df[(df['最新份额'] > 1.)]
df = df.sort_values(by=['基金折价率', '主力净流入-净额'], ascending=[False, False]).head(10)
display(HTML(df.to_html()))

## Caution: NEVER make investment decisions merely by looking at this indicator
```

## Rank by MFNIR

```python
df = ak.fund_etf_spot_em()
df["MFNIR"] = round(df["主力净流入-净额"] * abs(df["主力净流入-净占比"]) / 100.0, 2)
df = df.sort_values(by="MFNIR", ascending=False).head(20)
display(HTML(df.to_html()))
```
